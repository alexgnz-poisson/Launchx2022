Ejercicio 1: Trabajar con argumentos en funciones

Los argumentos requeridos en las funciones se utilizan cuando las funciones necesitan que esos argumentos funcionen correctamente. En este ejercicio, construirás un informe de combustible que requiere información de varias ubicaciones de combustible en todo el cohete.

Comienza por crear una función que necesite tres lecturas de combustible y devuelva un informe:



```python
# Función para leer 3 tanques de combustible y muestre el promedio
def generate_report(tanque_principal, tanque_externo, tanque_hidrogeno):
    total_average = (tanque_principal + tanque_externo + tanque_hidrogeno) / 3
    return f"""Fuel Report:
    Total Average: {total_average}%
    Tanque principal: {tanque_principal}%
    Tanque externo: {tanque_externo}%
    Tanque hidrogeno: {tanque_hidrogeno}% 
    """
```

Ahora que hemos definido la función de informes, vamos a comprobarlo. Para esta misión, los tanques no están llenos:

```python
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))
print(generate_report(80, 70, 85))
```

En lugar de simplemente crear el informe, la función también está calculando el promedio. Mejora la legibilidad extrayendo el cálculo promedio de la función en una nueva función para que el promedio se pueda hacer de forma independiente:

```python
# Función promedio 
def average(values):
    total = sum(values)
    number_of_items = len(values)
    return total / number_of_items

# Test the averaging function with a list of integers:
average([80, 85, 81]) 
```

```python
# Actualiza la función
def generate_report(tanque_principal, tanque_externo, tanque_hidrogeno):
    total_average = (tanque_principal + tanque_externo + tanque_hidrogeno) / 3
    return f"""Fuel Report:
    Total Average: {total_average}%
    Tanque principal: {tanque_principal}%
    Tanque externo: {tanque_externo}%
    Tanque hidrogeno: {tanque_hidrogeno}% 
    """

# Call the updated function again with different values
print(generate_report(88, 76, 70))
```

Ejercicio 2: Trabajo con argumentos de palabra clave

En este ejercicio, construirás un informe de cohete que requiere varias piezas de información, como el tiempo hasta el destino, el combustible a la izquierda y el nombre del destino. Comienza por crear una función que creará un informe preciso de la misión:

```python
# Función con un informe preciso de la misión. Considera hora de prelanzamiento, tiempo de vuelo, destino, tanque externo y tanque interno

def mission_report(pre_launch_time, hora_vuelo, destino, tanque_externo, tanque_principal):
    return f"""
    Mision hacia la {destino}
    Tiempo total del viaje: {pre_launch_time + hora_vuelo} minutos
    Total gasolina lado izquierdo: {tanque_externo + tanque_principal} galones
    """

print(mission_report(14, 51, "Luna", 200000, 300000))
```

La función es problemática porque no se puede adaptar para minutos adicionales o tanques adicionales de combustible. Hazlo más flexible permitiendo cualquier número de pasos basados en el tiempo y cualquier número de tanques. En lugar de usar *args y **kwargs, aprovecha el hecho de que puedes usar cualquier nombre que desees. Asegurate que la función sea más legible mediante el uso de nombres de variables que están asociados con las entradas, como *minutes y **fuel_reservoirs:

```python
# Escribe tu nueva función de reporte considerando lo anterior

def mission_report(destino, *minutos, **fuel_reservoirs):
    return f"""
    Mision hacia la {destino}
    Tiempo total de viaje: {sum(minutos)} minutes
    Total gasolina lado izquierdo: {sum(fuel_reservoirs.values())}
    """

print(mission_report("Luna", 10, 15, 51, main=300000, external=200000))
```

Debido a que el combustible que queda en los tanques es específico de cada tanque, actualiza la función para usar el nombre de cada tanque en el informe:

```python
# Escribe tu nueva función
def mission_report(destino, *minutos, **fuel_reservoirs):
    main_report = f"""
    Mision hacia la {destino}
    Tiempo total de viaje: {sum(minutos)} minutes
    Total gasolina lado izquierdo: {sum(fuel_reservoirs.values())}
    """
    for tank_name, gallons in fuel_reservoirs.items():
        main_report += f"{tank_name} tank --> {gallons} gallons left\n"
    return main_report

print(mission_report("Moon", 8, 11, 55, main=300000, external=200000))
```
